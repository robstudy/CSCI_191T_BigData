```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
md = pd.DataFrame(data2, columns=['movieId','rating'])
x = md.rating
sns.set(color_codes=True)
sns.distplot(x)
```

```python
sns.countplot(x='rating', data=md)
```

```python
r = {'0.5':0, '1.0':0,'1.5':0,'2.0':0,'2.5':0,'3.0':0,'3.5':0,'4.0':0,'4.5':0,'5.0':0}
for x in md.rating:
    if x == 0.5:
        r['0.5'] += 1
    elif x ==  1.0:
        r['1.0'] += 1
    elif x ==  1.5:
        r['1.5'] += 1
    elif x ==  2.0:
        r['2.0'] += 1
    elif x ==  2.5:
        r['2.5'] += 1
    elif x ==  3.0:
        r['3.0'] += 1
    elif x == 3.5:
        r['3.5'] += 1
    elif x == 4.0:
        r['4.0'] += 1
    elif x == 4.5:
        r['4.5'] += 1
    elif x == 5.0:
        r['5.0'] += 1
            
print(r)
```

```python
plt.title("Ratings")
plt.ylabel('Total')
plt.xlabel('Rating')
plt.bar(r.keys(), r.values(), 1.0, color='b')
```

```python
total = 0
for key in r:
    total += r[key] #get total
#print("Total:",total)
print('Percent Dispersion')
for key in r:
    print(key,':','%.2f' %( (r[key]/total)*100),'%')
```

```python
ratingLabels = []
ratingTotals = []
for key in r:
    ratingLabels.append(key)
    ratingTotals.append(r[key])
print(ratingLabels)
print(ratingTotals)
```

```python
plt.pie(ratingTotals, labels=ratingLabels, startangle=90, autopct='%.1f%%',radius=3.3)
plt.show()
```

```python
data3 = pd.read_csv('datasets/esea_master_kills_demos.part1.csv')
```

```python
csData = pd.DataFrame(data3, columns=['wp', 'wp_type'])
csData
```

```python
sns.countplot(x='wp', data=csData)
```

```python
cs_wep = []
for x in csData.wp:
    if x not in cs_wep:
        cs_wep.append(x)
        
cs_wep_num = []
for i in range(len(cs_wep)):
    cs_wep_num.append(0)
    
for x in csData.wp:
    cs_wep_num[cs_wep.index(x)] += 1
print(cs_wep)
print(cs_wep_num)
```

```python
plt.pie(cs_wep_num, labels=cs_wep, startangle=90, autopct='%.1f%%',radius=3.3)
plt.show()
```
